```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data=pd.read_excel('Downloads/Dataset_w6.xlsx')
data
```

# Find out how many countries are there in the dataset

```python
data.info()
```

```python
N=len(data['country'].value_counts())
print("Number_of_countries=", N)
```

# Examine the data by using visualizations


```python
plt.figure(figsize=(20,15))
sns.barplot(x='Region',y='GDP per capita (current US$)',data=data)
plt.title('Region VS GDP per capita (current US$)',fontsize=15)
plt.xticks(rotation = 45,fontsize = 15)
plt.show()
```

```python
plt.figure(figsize=(20,15))
sns.barplot(x='Region',y='International trade: Imports (million US$)',data=data)
plt.title('Region VS International trade: Imports (million US$)', fontsize=15)
plt.xticks(rotation = 45,fontsize = 15)
plt.show()
```

# Find out how many countries with respect to the region are availablein the dataset

```python
data['Region'].value_counts()
```

# GDP per capita (current US$) average

```python
data1=data.groupby("Region")['GDP per capita (current US$)'].mean().reset_index()
data1.sort_values('GDP per capita (current US$)',ascending=False)
```

```python
data1.nlargest(1,['GDP per capita (current US$)'])
```

# International trade: Imports (million US$)


```python
data2=data.groupby("Region")['International trade: Imports (million US$)'].mean().reset_index()
data2.sort_values('International trade: Imports (million US$)',ascending=False)
```

```python
data2.nlargest(1,['International trade: Imports (million US$)'])
```

# Find out which region has more consistent with respect to GDP percapita (current US$)

```python
data3=data.groupby("Region")['GDP per capita (current US$)'].std().reset_index()
data3.sort_values('GDP per capita (current US$)',ascending=False)
```
